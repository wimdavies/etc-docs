```python
### Search (search_profile=full)
```

Search with a full response contains:
    * info
    * records
    
Contained in 'info' is:
  * record_count - The number of matching records for the query
  * record_count_exact - Indication of if the above number is exact or an approximation
  * parameters - Reserved for future use
  * page_size - THe number of records returned in this page of results
  * pages - The total number of pages of results
  * page - The page of the current response
  
The 'records' field contains an array of object recordss matching the query. This is a subset of the full record

```python
import requests
req = requests.get('https://api.vam.ac.uk/v2/objects/search?q_material_technique=Jet&data_profile=full')
object_data = req.json()
print(object_data)
```
